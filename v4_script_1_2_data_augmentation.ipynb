```python
import pandas as pd
pd.set_option('display.max_rows', 200)
import numpy as np

import time
import random

import os
from os import getcwd 
from os.path import exists

version = 'v4'
update = True

getcwd() # current working directory
```

```python
load_name = f"{version}_data_all_sites.csv"
dataset = pd.read_csv(load_name)
protein_list = list(dataset.protein.unique())

print(f"total number of proteins: {len(protein_list)}")
display(dataset)
```

```python
match = [exists(f"./protein_charge/{x}") for x in protein_list]
dict_match = dict(zip(protein_list, match))

only_charge_folders = os.listdir('./protein_charge')
for name in protein_match:
    only_charge_folders.remove(name)
    
print("these proteins do not match with their names")
display(only_charge_folders)
print("whether the name matches or not")
display(dict_match)
```

```python
load_folder = './protein_charge'
charge_columns = pd.read_csv(f'{load_folder}/features-all-names.csv')

print("column names of charge-related dataset")
display(charge_columns.column_name)

charge_columns = list(charge_columns.column_name)
```

```python
protein_match = [name for name in protein_list if exists(f"./protein_charge/{name}")]
print("total number of matching proteins:", len(protein_match))

load_folder = './protein_charge'

save_folder = './augmented_dataset'
for name in protein_match:
    save_name = f'{version}_{name}.csv'
    if not exists(f'{save_folder}/{save_name}') or update:
        load_name = f'{load_folder}/{name}/all_AA/combined-output-all-residues-features.csv'
        charge_data = pd.read_csv(load_name, sep=' ', header=None)
        charge_data.columns = charge_columns[1:-2]
        charge_data = charge_data.sort_values(by='#').reset_index(drop=True)
        charge_data = charge_data.iloc[:,1:] # remove '#' column to prevent duplicated columns
        
        data = dataset[dataset['protein']==name].reset_index(drop=True)
        aug_data = pd.concat([data, charge_data], axis=1) 
        
        aug_data = aug_data.reset_index(drop=True)
        aug_data.to_pickle(f"{save_folder}/{save_name}.pkl")
        aug_data.to_csv(f"{save_folder}/{save_name}.csv", index=False)
    
    else:
        aug_data = pd.read_pickle(f"{save_name}.pkl")
    
display(aug_data)
```
